# Anharmonic phonon renormalisation with dynaphopy + GRACE (0.7 fs timestep)

Sibling notebook to `dynaphopy_grace_example.ipynb`, identical wiring but
with the MD timestep dropped from 1.0 fs to **0.7 fs** to match the
dynaphopy upstream Tersoff Si example exactly
(https://abelcarreras.github.io/DynaPhoPy/examples.html, which uses
`-ts 0.0007` ps). At a fixed `production_steps = 50 000` this gives
35 ps of trajectory (vs the 50 ps the 1.0 fs sibling produces); a finer
integrator step trades a bit of total simulation time for better
numerical accuracy and a higher Nyquist limit in the power spectrum.

Walks through `calculate_phonon_md_renormalisation` (v0.0.8) end-to-end
on Si 2×2×2 with the **GRACE-1L-OAM** foundation model. Shows two equivalent paths:

1. **ASE-driven** — Langevin NVT MD via ASE's built-in integrator + GRACE calculator,
   routed through `calculate_phonon_md_renormalisation` (the user-facing macro).
2. **LAMMPS-driven** — same FM via LAMMPS' `pair_style grace`, NVT MD via LAMMPS' own
   integrator, trajectory parsed by `dynaphopy.interface.iofile` and fed into the
   same `_project_with_dynaphopy` node.

References:
- dynaphopy upstream examples: <https://abelcarreras.github.io/DynaPhoPy/examples.html>
- GRACE foundation models: <https://github.com/ICAMS/grace-tensorpotential>

**Hardware notes:** GRACE runs on GPU here (RTX 5070 Ti, sm_120). On Blackwell
GPUs the nvidia-* CUDA wheels bundled with TF 2.20 must be upgraded
(`nvidia-cuda-nvcc-cu12>=12.8`, `nvidia-cudnn-cu12>=9.22`, etc.) to support
ptxas compilation for compute capability 12.0. The first XLA-compile step on
a fresh process takes ~10 s; subsequent evals are ~25 ms/step on this hardware.

## 1. Structure + GRACE foundation model

Si primitive (diamond, 2-atom basis) → 2×2×2 supercell for MD (16 atoms).
We seed the EOS scan with the textbook value `a = 5.43 Å`, but the
GRACE-optimised `a0` (refit in Section A by `optimise_cubic_lattice_parameter`)
is what both halves of the notebook actually use — the textbook value is
only the initial guess and a sanity-check anchor.

```python
# Initial-guess primitive at the textbook Si lattice constant (5.43 Å). The
# GRACE-1L-OAM equilibrium will be re-derived in Section A by an EOS scan +
# Birch–Murnaghan fit (`optimise_cubic_lattice_parameter`), and `si_supercell`
# is rebuilt below from that optimised primitive so Section A's MD and
# Section B's LAMMPS run share the same GRACE-equilibrated geometry. The
# `a=5.43` value here is only the starting point for the EOS scan and the
# smoke-test single-point evaluation in the next cell.
si_prim_initial = bulk("Si", "diamond", a=5.43)
fc2_multiplier = 2 * np.eye(3, dtype=int)
print(f"Si primitive (initial guess @ a=5.43 Å): {len(si_prim_initial)} atoms, "
      f"cell={si_prim_initial.cell.lengths()}")
print(f"FC2 multiplier diagonal: {fc2_multiplier.diagonal().tolist()} "
      f"→ {len(si_prim_initial) * int(round(np.linalg.det(fc2_multiplier)))}-atom supercell")
```

```python
from tensorpotential.calculator.foundation_models import grace_fm

grace_calc = grace_fm("GRACE-1L-OAM")
si_prim_check = si_prim_initial.copy()
si_prim_check.calc = grace_calc
e0 = si_prim_check.get_potential_energy()
print(f"GRACE-1L-OAM on Si primitive (a=5.43 Å): "
      f"E = {e0:.4f} eV  ({e0/len(si_prim_check):.4f} eV/atom)")
```

## 2. Section A — ASE-driven workflow through `calculate_phonon_md_renormalisation`

This section is wired as a single `pyiron_workflow.Workflow` so that
upstream-vs-downstream dependencies are explicit. The graph has two nodes:

1. **`wf.lat_opt = optimise_cubic_lattice_parameter(...)`** — a cubic EOS
   scan (`strain_range=(-0.02, 0.02)`, `num_points=5`) followed by a
   Birch–Murnaghan fit. This re-derives the GRACE-1L-OAM equilibrium
   lattice parameter `a0`, which can drift from the textbook value
   (`a=5.43 Å`) by a fraction of a percent. We feed that optimised
   geometry into both the ASE phonon-MD path (here) and the LAMMPS
   path (Section B) — apples-to-apples.
2. **`wf.phonon = calculate_phonon_md_renormalisation(...)`** — the
   user-facing macro. It internally:
   1. resolves arguments (`_resolve_md_defaults`),
   2. fits FC2 from displacement-force evals through `engine.calculate`
      (`_compute_fc2_from_scratch`),
   3. runs ASE Langevin NVT MD borrowing the engine's calculator
      (`_run_nvt_trajectory`),
   4. builds `dynaphopy.Quasiparticle` and fits Lorentzians
      (`_project_with_dynaphopy`).

   Returns a typed `MdPhononOutput` dataclass with harmonic and
   renormalised frequencies, per-band power spectra, linewidths, MD
   diagnostics, and optional handles to the underlying dynaphopy /
   phonopy objects.

We use two engines:
- `engine_min` (`CalcInputMinimize`, `force_convergence_tolerance=0.05`) — drives
  the per-strain relaxations inside the EOS scan.
- `ase_engine` (`CalcInputStatic`) — drives FC2 displacements and supplies
  the calculator the macro borrows for Langevin NVT MD.

Both engines wrap the same `grace_calc`; `engine.with_working_directory(...)`
keeps the on-disk artefacts separated.

```python
from pyiron_workflow import Workflow
from pyiron_workflow_atomistics.engine import (
    ASEEngine,
    CalcInputMinimize,
    CalcInputStatic,
)
from pyiron_workflow_atomistics.structure import get_bulk
from pyiron_workflow_atomistics.physics.bulk import optimise_cubic_lattice_parameter
from pyiron_workflow_atomistics.physics.phonons import (
    calculate_phonon_md_renormalisation,
)

# Engine for the per-strain relaxations inside the EOS scan. CalcInputMinimize
# triggers a BFGS minimisation under GRACE forces at each scanned volume.
engine_min = ASEEngine(
    EngineInput=CalcInputMinimize(
        force_convergence_tolerance=0.05, max_iterations=100
    ),
    calculator=grace_calc,
    working_directory="./_dynaphopy_grace_long_runs",
)

# Engine for FC2 displacement-force evaluations and the Langevin NVT MD
# trajectory generator borrowed by `calculate_phonon_md_renormalisation`.
ase_engine = ASEEngine(
    EngineInput=CalcInputStatic(),
    calculator=grace_calc,
    working_directory="./_dynaphopy_grace_long_runs",
)

# Band-path of commensurate q-points for the 2×2×2 supercell — these
# are the only q's where dynaphopy's mode projection is well-defined.
# Γ → X → K/W-like → L is a standard FCC path through high-symmetry points.
q_points_commensurate = [
    [0.0, 0.0, 0.0],
    [0.5, 0.0, 0.0],
    [0.5, 0.5, 0.0],
    [0.5, 0.5, 0.5],
]

wf_ase = Workflow("dynaphopy_grace_ase")

# Node 1: GRACE-driven cubic lattice-parameter optimisation. Returns the
# 8-atom *cubic* equilibrium structure plus the scalar `a0` (in Å).
wf_ase.lat_opt = optimise_cubic_lattice_parameter(
    structure=si_prim_initial,
    name="Si",
    crystalstructure="diamond",
    engine=engine_min.with_working_directory("lat_opt"),
    strain_range=(-0.02, 0.02),
    num_points=5,
)

# Node 2: rebuild the 2-atom *primitive* at the optimised lattice parameter.
# `optimise_cubic_lattice_parameter` returns its `equil_struct` as the
# conventional cubic cell (it calls `get_bulk(..., cubic=True)` internally).
# Phonon-MD and the LAMMPS half use the 2-atom primitive × 2×2×2 supercell
# (16 atoms), so we rebuild the primitive from `a0` here.
wf_ase.prim_opt = get_bulk(
    name="Si",
    crystalstructure="diamond",
    a=wf_ase.lat_opt.outputs.a0,
    cubic=False,
)

# Node 3: the phonon-MD macro on the GRACE-equilibrated primitive.
wf_ase.phonon = calculate_phonon_md_renormalisation(
    structure=wf_ase.prim_opt.outputs.equil_struct,
    engine=ase_engine.with_working_directory("phonon"),
    fc2_supercell_matrix=fc2_multiplier,
    temperature=788.0,
    equilibration_steps=5000,
    production_steps=50000,
    time_step=0.7,
    thermostat_time_constant=100.0,
    q_points=q_points_commensurate,
    seed=42,
    keep_handles=True,
    power_spectra=True,
)

# GRACE's TF FuncGraph isn't pickleable; disable pyiron_workflow's recovery save.
wf_ase.recovery = None
wf_ase.run()

# Surface the optimised lattice parameter early — this is the geometry both
# halves of the notebook now share.
a0_grace = float(wf_ase.lat_opt.outputs.a0.value)
B_grace = float(wf_ase.lat_opt.outputs.B.value)
e0_per_atom = float(wf_ase.lat_opt.outputs.equil_energy_per_atom.value)
print(f"GRACE-1L-OAM optimised a0 = {a0_grace:.4f} Å "
      f"(vs textbook 5.43 Å, Δ = {a0_grace - 5.43:+.4f} Å)")
print(f"GRACE-1L-OAM bulk modulus B = {B_grace:.1f} GPa")
print(f"GRACE-1L-OAM e0/atom       = {e0_per_atom:.4f} eV")

out_ase = wf_ase.phonon.outputs.md_phonon_output.value
si_prim = wf_ase.prim_opt.outputs.equil_struct.value  # optimised 2-atom primitive

print(f"⟨T⟩ measured: {out_ase.md_temperature_mean:.1f} K  "
      f"(target {out_ase.temperature:.0f} K)")
print(f"σ_T: {out_ase.md_temperature_std:.1f} K")
print(f"harmonic   freqs (THz): {out_ase.harmonic_frequencies[0]}")
print(f"renormalised freqs (THz): {out_ase.renormalised_frequencies[0]}")
```

### Section A plot — renormalised phonon dispersion (ASE) + power spectrum at X

Left: harmonic dispersion (continuous lines from phonopy at each commensurate
q-point) vs MD-projected renormalised frequencies (markers, dynaphopy fit on
the Langevin trajectory). Right: per-band power spectra at q = X, with
vertical dotted lines marking the harmonic positions.

```python
# Smooth harmonic dispersion: sample phonopy along Γ→X→K/W→L at ~50 pts per
# segment, plot as continuous curves. Renormalised MD-projected frequencies
# (4 commensurate q's) are overlaid as markers at the corresponding segment-
# length positions. Cached as `x_path / freqs_path / x_renorm / q_path_labels`
# so Section 4 (side-by-side ASE vs LAMMPS) can reuse them without re-querying
# phonopy.
def _fine_band_path(phonopy_view, n_pts_per_segment=51):
    corners = [(0.0, 0.0, 0.0), (0.5, 0.0, 0.0),
               (0.5, 0.5, 0.0), (0.5, 0.5, 0.5)]
    labels_ = ["Γ", "X", "K/W", "L"]
    qs_list, x_list = [], []
    cum, x_corners_ = 0.0, [0.0]
    for i in range(len(corners) - 1):
        q0 = np.array(corners[i], dtype=float)
        q1 = np.array(corners[i + 1], dtype=float)
        seg_len = float(np.linalg.norm(q1 - q0))
        ts = np.linspace(0.0, 1.0, n_pts_per_segment,
                         endpoint=(i == len(corners) - 2))
        qs_list.append(q0 + ts[:, None] * (q1 - q0))
        x_list.append(cum + ts * seg_len)
        cum += seg_len
        x_corners_.append(cum)
    qs = np.vstack(qs_list)
    x = np.concatenate(x_list)
    freqs = np.array([phonopy_view.get_frequencies(q) for q in qs])
    return x, freqs, x_corners_, labels_

x_path, freqs_path, x_commensurate, q_path_labels = _fine_band_path(out_ase.phonopy)
x_renorm = np.asarray(x_commensurate)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 4))

for b in range(freqs_path.shape[1]):
    ax1.plot(x_path, freqs_path[:, b],
             color=f"C{b}", lw=2, alpha=0.6,
             label="harmonic" if b == 0 else None)
    ax1.scatter(x_renorm, out_ase.renormalised_frequencies[:, b],
                color=f"C{b}", marker="o", s=50, edgecolor="black", linewidth=0.5,
                label="renorm (788 K)" if b == 0 else None)
ax1.set_xticks(x_renorm)
ax1.set_xticklabels(q_path_labels)
ax1.set_xlabel("q-point")
ax1.set_ylabel("frequency (THz)")
ax1.set_title("Renormalised phonon bands — Si 2×2×2 (GRACE-1L-OAM, ASE)")
ax1.legend()
ax1.grid(alpha=0.3)

# Power spectrum at X (q-index 1).
i_X = 1
for b in range(out_ase.power_spectra.shape[1]):
    ax2.plot(out_ase.frequency_grid, out_ase.power_spectra[i_X, b],
             label=f"band {b}", alpha=0.85)
    ax2.axvline(out_ase.harmonic_frequencies[i_X, b],
                ls=":", color=f"C{b}", alpha=0.4)
ax2.set_xlim(0, 30)
ax2.set_xlabel("frequency (THz)")
ax2.set_ylabel("power (arb.)")
ax2.set_title("MD-projected power spectrum at X — ASE\n(dotted lines = harmonic)")
ax2.legend(ncols=3, fontsize=8)
ax2.grid(alpha=0.3)
fig.tight_layout()
plt.show()
```

## 3. Section B — LAMMPS-driven workflow with `pair_style grace`

Same foundation model, different MD driver. Pieces:
- **`LammpsEngine`** from `pyiron_workflow_lammps` wraps the `lmp` binary at
  `/home/liger/lammps/build/lmp` (built from the `grace` branch with ML-PACE/ML-GRACE).
- For trajectory generation we drive LAMMPS directly (its own integrator is much
  faster than per-step ASE subprocess calls) and dump positions+velocities.
- `dynaphopy.interface.iofile.read_lammps_trajectory` parses the dump and
  builds a `dynaphopy.dynamics.Dynamics` object compatible with `Quasiparticle`.
- FC2 is reused from Section A so both halves are fit against the same harmonic reference.
- **Same geometry as Section A.** The 2×2×2 supercell is rebuilt below from
  Section A's GRACE-optimised primitive (`a0 = {a0_grace}` Å, surfaced from
  `wf_ase.lat_opt`), so this LAMMPS path and the ASE path differ *only* in
  MD driver — never in lattice parameter, supercell tiling, or FC2 reference.

```python
import subprocess
from pathlib import Path

lammps_workdir = Path("./_dynaphopy_grace_long_runs/lammps_run").resolve()
lammps_workdir.mkdir(parents=True, exist_ok=True)
LMP_BIN = "/home/liger/lammps/build/lmp"
GRACE_MODEL_DIR = "/home/liger/.cache/grace/GRACE-1L-OAM"

# Rebuild the 2×2×2 Si supercell from the GRACE-optimised primitive so
# Section B's LAMMPS MD runs on the same geometry Section A used for FC2 +
# ASE Langevin MD. `si_prim` was bound above to `wf_ase.prim_opt.outputs.equil_struct`.
si_supercell = make_supercell(si_prim, fc2_multiplier)
print(f"Si primitive (GRACE-equilibrated, a0={a0_grace:.4f} Å): "
      f"{len(si_prim)} atoms, cell={si_prim.cell.lengths()}")
print(f"2x2x2 supercell: {len(si_supercell)} atoms, "
      f"cell={si_supercell.cell.lengths()}")

# Write Si 2x2x2 LAMMPS data file.
data_file = lammps_workdir / "si_222.data"
from ase.io import write as ase_write
ase_write(str(data_file), si_supercell, format="lammps-data", masses=True)
print(f"Wrote {data_file} ({len(si_supercell)} atoms)")
```

```python
# LAMMPS NVT-Langevin input file. Same temperature, thermostat coupling, and
# step count as Section A, written to a dump file for dynaphopy ingestion.
lammps_input = lammps_workdir / "nvt.in"
lammps_dump  = lammps_workdir / "trajectory.lammpstrj"
lammps_log   = lammps_workdir / "log.lammps"
lammps_input.write_text(f"""\
units           metal
atom_style      atomic
boundary        p p p
read_data       {data_file}

pair_style      grace
pair_coeff      * * {GRACE_MODEL_DIR} Si

neighbor        2.0 bin
neigh_modify    every 1 delay 0 check yes

# Equilibration: Langevin NVT for 3.5 ps (5000 steps × 0.7 fs).
velocity        all create 788.0 42 dist gaussian
fix             1 all nve
fix             2 all langevin 788.0 788.0 0.1 42
timestep        0.0007                                # ps (= 0.7 fs / 1000)
thermo          500
thermo_style    custom step temp pe etotal
run             5000

# Production — dump positions and velocities every step for the MD-projection.
unfix 1
unfix 2
fix             1 all nve
fix             2 all langevin 788.0 788.0 0.1 43
reset_timestep  0
dump            1 all custom 1 {lammps_dump} id type xu yu zu vx vy vz
dump_modify     1 sort id
run             50000
""")

env = os.environ.copy()
env["LD_LIBRARY_PATH"] = "/usr/lib/wsl/lib:" + env.get("LD_LIBRARY_PATH", "")
print("Running LAMMPS NVT with pair_style grace (GPU)...")
result = subprocess.run(
    [LMP_BIN, "-in", str(lammps_input), "-log", str(lammps_log)],
    cwd=str(lammps_workdir),
    env=env,
    capture_output=True, text=True, timeout=3600,
)
print(f"return code: {result.returncode}")
tail_lines = result.stdout.splitlines()[-10:]
for line in tail_lines:
    print(line)
```

```python
# Parse the LAMMPS dump and assemble a trajectory_pack compatible with
# _project_with_dynaphopy. We replicate _run_nvt_trajectory's output schema
# from the dump file so the same downstream synthesis node can ingest it.

from pyiron_workflow_atomistics.physics.phonons.md_renormalised import (
    _multiplier_to_cell_vectors,
    _project_with_dynaphopy,
)

def read_lammps_custom_dump(dump_path: Path):
    """Parse a LAMMPS 'dump custom id type xu yu zu vx vy vz' file.

    Returns dict with positions (n_step, n_atom, 3), velocities (same shape),
    n_atom, n_steps. Time is reconstructed from the LAMMPS step count.
    """
    text = dump_path.read_text().splitlines()
    i = 0
    n_atom = None
    positions_list, velocities_list = [], []
    while i < len(text):
        if text[i].startswith("ITEM: TIMESTEP"):
            i += 2
        elif text[i].startswith("ITEM: NUMBER OF ATOMS"):
            n_atom = int(text[i + 1].strip())
            i += 2
        elif text[i].startswith("ITEM: BOX BOUNDS"):
            i += 4
        elif text[i].startswith("ITEM: ATOMS"):
            header = text[i].split()[2:]
            col = {name: idx for idx, name in enumerate(header)}
            frame_pos = np.zeros((n_atom, 3))
            frame_vel = np.zeros((n_atom, 3))
            for k in range(n_atom):
                row = text[i + 1 + k].split()
                aid = int(row[col["id"]]) - 1
                frame_pos[aid] = [float(row[col[c]]) for c in ("xu", "yu", "zu")]
                frame_vel[aid] = [float(row[col[c]]) for c in ("vx", "vy", "vz")]
            positions_list.append(frame_pos)
            velocities_list.append(frame_vel)
            i += 1 + n_atom
        else:
            i += 1
    pos = np.stack(positions_list, axis=0)
    vel = np.stack(velocities_list, axis=0)
    return pos, vel

lammps_pos, lammps_vel = read_lammps_custom_dump(lammps_dump)
print(f"Parsed dump: positions {lammps_pos.shape}, velocities {lammps_vel.shape}")
```

```python
# Build the trajectory_pack and feed into _project_with_dynaphopy.
# Reuses Section A's FC2 (identical primitive + FC2 supercell), so both halves
# share the same harmonic reference and we're isolating MD-driver differences.

time_step_fs = 0.7
n_steps = lammps_pos.shape[0]
# ⟨T⟩ and σ_T are read directly from LAMMPS' thermo output (log.lammps);
# velocity-units bookkeeping is left to LAMMPS' own thermostat.

log_text = lammps_log.read_text()
# LAMMPS thermo header is " Step  Temp  PotEng  TotEng" (note leading whitespace).
# Walk the log block-by-block: every header line opens a thermo block, the next
# whitespace-only block of numeric lines is the data. Production = second block.
prod_temps = []
blocks: list[list[float]] = []
current: list[float] = []
in_block = False
for raw in log_text.splitlines():
    line = raw.strip()
    if line.startswith("Step") and "Temp" in line:
        if current:
            blocks.append(current)
            current = []
        in_block = True
        continue
    if in_block:
        toks = line.split()
        if len(toks) >= 2 and toks[0].isdigit():
            try:
                current.append(float(toks[1]))
            except ValueError:
                pass
        elif line.startswith("Loop") or line.startswith("ERROR"):
            blocks.append(current)
            current = []
            in_block = False
if current:
    blocks.append(current)

prod_temps = blocks[1] if len(blocks) >= 2 else (blocks[0] if blocks else [])
md_T_mean = float(np.mean(prod_temps)) if prod_temps else float("nan")
md_T_std  = float(np.std(prod_temps))  if prod_temps else float("nan")
print(f"LAMMPS production ⟨T⟩: {md_T_mean:.1f} K  (target 788 K)")
print(f"LAMMPS production σ_T: {md_T_std:.1f} K")

from pyiron_workflow_atomistics.physics.phonons.md_renormalised import (
    _multiplier_to_cell_vectors,
    _project_with_dynaphopy,
)

trajectory_pack_lammps = {
    "positions": lammps_pos,
    "velocities": lammps_vel,            # carried for reference; not used downstream
    "time": np.arange(n_steps) * time_step_fs,                                # fs
    "supercell": _multiplier_to_cell_vectors(si_prim.cell, fc2_multiplier),   # 3x3 cell vectors
    "n_md_steps": n_steps,
    "time_step_fs": time_step_fs,
    "md_temperature_mean": md_T_mean,
    "md_temperature_std": md_T_std,
}

# Reuse FC2 from Section A's macro output (keep_handles=True kept the phonopy view).
fc2_reused = np.asarray(out_ase.phonopy.force_constants)

# Project via the macro's _project_with_dynaphopy node directly (now unit-clean
# upstream — `velocity=None` is set inside the node). Same physics as Section A's
# projection step, fed a LAMMPS-generated trajectory.
out_lammps = _project_with_dynaphopy.node_function(
    structure=si_prim,
    fc2_array=fc2_reused,
    resolved_fc2_supercell=fc2_multiplier,
    trajectory_pack=trajectory_pack_lammps,
    resolved_q_points=np.array(
        [[0.0, 0.0, 0.0], [0.5, 0.0, 0.0], [0.5, 0.5, 0.0], [0.5, 0.5, 0.5]]
    ),
    temperature=788.0,
    power_spectra=True,
    keep_handles=False,
)
print(f"LAMMPS converged: {out_lammps.converged}")
print(f"renormalised freqs (THz): {out_lammps.renormalised_frequencies[0]}")
```

## 4. Side-by-side comparison — renormalised phonon bands

Both halves share the same FC2 (from Section A), the same primitive, the
same supercell, the same target T = 788 K, and the same production step
count (50 000). Differences in the renormalised frequencies isolate
**MD-driver differences** (ASE Langevin vs LAMMPS Langevin) and the
underlying trajectory sampling.

## 5. Maxwell-Boltzmann velocity-distribution analysis

Sanity-check the MD thermostat by histogramming the per-atom **mass-weighted**
velocity magnitudes (`v · √m_amu`, in √u·Å/ps) from the LAMMPS production
trajectory and fitting a Maxwell distribution. The scale parameter `a` from
`scipy.stats.maxwell.fit` recovers `T = a² / kB` with
`kB = 0.831446 u·Å²/(ps²·K)` — this is exactly the convention used by
dynaphopy's own `boltzmann_distribution` analyser
(`dynaphopy/analysis/energy.py`), so the histogram matches the numbers on
the upstream Si Tersoff example page (⟨|v|⟩_mw ≈ 40.7 √u·Å/ps at T ≈ 788 K).


```python
from scipy.stats import maxwell

# dynaphopy's `boltzmann_distribution` (dynaphopy/analysis/energy.py) uses the
# mass-weighted velocity v_mw[i,a] = v[i,a] * sqrt(m_a) in √u · Å/ps, and
# back-derives the temperature from the Maxwell-Boltzmann scale parameter via
#     T = a^2 / kB     with kB = 0.831446 u·Å²/(ps²·K).
# We reproduce that convention here so the histogram aligns with the numbers
# on the upstream dynaphopy Si Tersoff example page (⟨|v|⟩_mw ≈ 40.7 √u·Å/ps
# at T ≈ 788 K).
masses_amu = np.asarray(si_supercell.get_masses())          # shape (n_atoms,)
v_mw = lammps_vel * np.sqrt(masses_amu)[None, :, None]      # √u·Å/ps
v_mag_mw = np.linalg.norm(v_mw, axis=2).ravel()
v_mean = float(v_mag_mw.mean())
v_std  = float(v_mag_mw.std())

kB_dynaphopy = 0.831446  # u·Å² / (ps² · K)
fit_params = maxwell.fit(v_mag_mw, floc=0, scale=np.mean([v_mean, v_std]))
a_fit = float(fit_params[1])
T_inferred = a_fit**2 / kB_dynaphopy

print(f"⟨|v|⟩ (mass-weighted) = {v_mean:.3f} √u·Å/ps")
print(f"σ_|v|                = {v_std:.3f} √u·Å/ps")
print(f"Maxwell fit a        = {a_fit:.3f} √u·Å/ps")
print(f"T from a²/kB         = {T_inferred:.1f} K   (target 788 K)")

# Analytical Maxwell-Boltzmann PDF at the fitted scale parameter.
v_grid = np.linspace(0.0, v_mag_mw.max() * 1.05, 400)
mb_pdf = maxwell.pdf(v_grid, loc=0.0, scale=a_fit)

fig, ax = plt.subplots(figsize=(7, 4))
ax.hist(v_mag_mw, bins=80, density=True, alpha=0.6, color="C0",
        label="LAMMPS MD (mass-weighted)")
ax.plot(v_grid, mb_pdf, color="C3", lw=2,
        label=f"Maxwell fit @ T = {T_inferred:.0f} K")
ax.set_xlabel(r"$|v|\sqrt{m}$  ($\sqrt{u}\cdot$Å/ps)")
ax.set_ylabel("probability density")
ax.set_title("Per-atom mass-weighted velocity distribution — LAMMPS production "
             "(dynaphopy convention)")
ax.legend()
ax.grid(alpha=0.3)
fig.tight_layout()
plt.show()
```

## 6. Notes & caveats

- 50 000 production steps at 1 fs (50 ps trajectory) + 5 ps equilibration
  sits roughly 7× below the dynaphopy upstream Tersoff Si reference
  sample budget (500 000 × 0.7 fs ≈ 350 ps), which is enough to get
  well-separated optical peaks but still leaves visible width in the
  per-mode Lorentzian fits. Further extending `production_steps`
  toward ~200–500 k continues to tighten the fits at a roughly linear
  cost of GPU time (~4 min per additional 10 k steps on this hardware).
- At Γ on monatomic primitives, the three acoustic modes are degenerate
  at zero — dynaphopy's Lorentzian fit becomes ill-conditioned. The notebook
  demonstrates the workflow at q = X (0.5, 0, 0) where all 6 modes are
  non-degenerate. For Γ-point work, ensure the cell has a ≥ 2-atom basis
  and consider longer trajectories.
- Scale verified against the dynaphopy upstream Tersoff Si example
  (https://abelcarreras.github.io/DynaPhoPy/examples.html): Tersoff Si
  at 788 K shows ~0.5 THz frequency shifts (~3 %) on the ~15 THz optical
  mode. Our notebook on the same primitive (different FM, different T)
  should show shifts of similar order — values that look like a uniform
  ratio across all bands point at a units / convention issue rather than
  anharmonicity. (One such bug — feeding ASE-internal velocity values
  in `Å/AU_t` to dynaphopy as if they were `Å/ps` — was found and fixed
  while writing this notebook; the macro now passes `velocity=None` so
  dynaphopy reconstructs the velocity field cleanly from positions.)
- The macro auto-warns when `⟨T⟩` drifts by more than 3 % from the
  requested temperature. Short MD on small cells tends to overshoot
  the tolerance — this is expected and is documented in the macro's
  `MdPhononOutput.check_md_health()` method.
- For polar materials (LO–TO splitting), pass `born_charges` and
  `epsilon_inf` once the v2 BORN-correction follow-up lands. The
  `[phonons-md]` install extra already pulls in dynaphopy for this.
- To replace GRACE with another foundation model: swap `grace_fm(...)`
  for any ASE-compatible calculator (MACE, M3GNet, CHGNet, …) in
  Section A; for Section B, point `pair_style` and `pair_coeff` at
  the equivalent LAMMPS pair style.
