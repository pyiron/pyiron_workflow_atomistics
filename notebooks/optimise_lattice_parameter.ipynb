# Cubic lattice-parameter optimisation

Wraps `eos_volume_scan` and an EOS fit to return the equilibrium
cubic lattice parameter, bulk modulus, energy, and equilibrium
structure for an FCC Cu crystal under the ASE EMT potential.


```python
from ase.calculators.emt import EMT

from pyiron_workflow_atomistics.engine import ASEEngine, CalcInputMinimize
from pyiron_workflow_atomistics.structure import get_bulk
from pyiron_workflow_atomistics.physics.bulk import optimise_cubic_lattice_parameter

engine = ASEEngine(
    EngineInput=CalcInputMinimize(force_convergence_tolerance=0.05),
    calculator=EMT(),
    working_directory="./_optlat_runs",
)
```

```python
structure = get_bulk.node_function("Cu", crystalstructure="fcc", a=3.55, cubic=True)

wf = optimise_cubic_lattice_parameter(
    structure=structure,
    name="Cu",
    crystalstructure="fcc",
    engine=engine,
    rattle_amount=0.0,
    strain_range=(-0.02, 0.02),
    num_points=5,
)
wf.run()

print(f"a0    = {wf.outputs.a0.value:.4f} A")
print(f"B     = {wf.outputs.B.value:.1f} GPa")
print(f"e0/at = {wf.outputs.equil_energy_per_atom.value:.4f} eV")
print(f"v0/at = {wf.outputs.equil_volume_per_atom.value:.3f} A^3")
```
